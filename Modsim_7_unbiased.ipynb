```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def random_walk(n_iter,n_step):
      all_walks=[]
      all_iter=[]
      for k in range(n_iter):
        it=[]
        seq=[]
        k=0
        dist=0
        r=np.random.rand(n_step)
        for i in r:
          if(i<=0.5): #probability of having tail = 0.5
            dist+=-1
            seq=np.append(seq,dist)
          else:
            dist+=1
            seq=np.append(seq,dist)
          it=np.append(it,k)
          k+=1
        all_iter=np.append(all_iter,it)
        all_walks=np.append(all_walks,seq)
        all_walks=all_walks.reshape(-1,15)
      return all_walks
```

```python
import random

def random_walk(steps):
    position = 0
    positions = [position]

    for _ in range(steps):
        if random.random() < 0.5:
            position += 1
        else:
            position -= 1
        positions.append(position)

    return positions

num_trials = 1000  # Number of trials to obtain statistics
num_steps = 100    # Number of time steps in each trial

expected_positions = [0] * (num_steps + 1)

for _ in range(num_trials):
    positions = random_walk(num_steps)
    expected_positions = [x + y for x, y in zip(expected_positions, positions)]

expected_positions = [x / num_trials for x in expected_positions]

print("Time Step\tExpected Position")
for t, pos in enumerate(expected_positions):
    print(f"{t}\t\t{pos}")
```

```python
def formean(n):
    allmean=[]
    walks=random_walk(n,15)
    column_sums = np.sum(walks, axis=0)
    Mean=column_sums/n
    allmean=np.append(allmean,Mean)
    return allmean
```

```python
r=range(15)
matrix=formean(50)
matrix1=formean(500)
matrix2=formean(5000)
plt.plot(r,matrix,label='50 iterations')
plt.plot(r,matrix1,label='500 iterations')
plt.plot(r,matrix2,label='5000 iterations')
plt.xlabel('Steps')
plt.ylabel('Mean position at each step(for fixed iterations)')
plt.legend()
```

```python
column_variances = np.var(random_walk(50,15), axis=0)
plt.plot(r,column_variances,label='50 iterations')
column_variances1 = np.var(random_walk(500,15), axis=0)
plt.plot(r,column_variances1,label='500 iterations')
column_variances2 = np.var(random_walk(5000,15), axis=0)
plt.plot(r,column_variances2,label='5000 iterations')
plt.xlabel('Steps')
plt.ylabel('Variances for each step(for fixed number of iterations)')
plt.legend()
```
