## ABHI TAK T1, T2 se analyse kar diya ab A k effect ki baat karte

As of A we would expect a sudden decrease in the number of susceptibles available with increase in A(effectivity of lockdown). If increased effective lockdown is implemented in the begining then the peak is delayed more however we might expect that the number of infected at peak will be higher because we restricted the pool of susceptibles before the peak more than the previous case so after the lockdown the pool will be larger than the previous case and will cause a larger peak. If a lockdown with increased restrictions is implemented during the time when peak is going to occur then the peak would be shifted to the right, later, as well as the peak will be less in terms of the number of people infected.

```python
#Assume that the lockdown starts after 10 days

sus=[7620]
s=7619
N=7620
infected=[6]
inf=6
x=1/2
time=[0]
rec=[0]
re=0
a=0.1
dt=1
A=0.1
for i in np.arange(dt,160,dt):
    if(i>5 and i<100):
      if(A<1):
        A+=0.05
      else:
        A=1
      pr=(x)*(1-A)
    else:
      pr=x
    s=s-(pr)*inf*s/N*(dt)
    sus.append(s)
    inf=inf + (pr)*s*inf/N*(dt)-(a)*inf*(dt)
    infected.append(inf)
    re=re+(a)*inf*(dt)
    rec.append(re)
    time.append(i)
max_inf=max(infected)
print(max_inf)
index = infected.index(max_inf)
print(time[index])
plt.plot(time,sus,label='SUS')
plt.plot(time,infected,label='INF')
plt.plot(time,rec,label='REC')
plt.legend()
# 1/2 - value of A, max_infected,day of the peak in infeected
# 0.3 - 24595.53208236888,41
# 0.5 - 16825.34828873242,55
# 0.7 - 26907.68997912188,92

# 2/3
# 0.3 - 30047.73332877433 , 29
# 0.5 -23627.564754054118 , 37
# 0.7 - 11463.179994770802, 59
#26782.93615992235
#28
```

In this case we observe that there will be two peaks, one when the the lockdown in running which will be relatively quiet small and will occur soon before the lockdown is hundred percent efficient because after that point there will be no susceptibles after that point and the number of infected will decrease to a constant value however after the lockdown is relased we will observe a new peak after some days because the large pool of susceptibles is only now released for the infecteds to infect so it will be like the normal SIR model. By this effect we have just delayed the peak so that meanwhile we can prepare or work on creating medicines and vaccines to fight the epidemic.
