```python
#Assume that the lockdown starts after 10 days

sus=[76200]
s=76194
N=76200
infected=[6]
inf=6
x=1/2
time=[0]
rec=[0]
re=0
a=0.1
dt=1
A=0.5
for i in np.arange(dt,100,dt):
    if(i>20 and i<40):
      pr=(x)*(1-A)
    else:
      pr=x
    s=s-(pr)*inf*s/N*(dt)
    sus.append(s)
    inf=inf + (pr)*s*inf/N*(dt)-(a)*inf*(dt)
    infected.append(inf)
    re=re+(a)*inf*(dt)
    rec.append(re)
    time.append(i)
max_inf=max(infected)
print(max_inf)
index = infected.index(max_inf)
print(time[index])
plt.plot(time,sus,label='SUS')
plt.plot(time,infected,label='INF')
plt.plot(time,rec,label='REC')
plt.legend()
# 1/2 - value of A, max_infected,day of the peak in infeected
# 0.3 - 24595.53208236888,41
# 0.5 - 16825.34828873242,55
# 0.7 - 26907.68997912188,92

# 2/3
# 0.3 - 30047.73332877433 , 29
# 0.5 -23627.564754054118 , 37
# 0.7 - 11463.179994770802, 59
#26782.93615992235
#28
```

To solve the given problem we break the vaccinated compartment into three compartments as the problem can be understood in reference to three days. If we understand the dynamics around three days then we can extend the model further to any number of days as the behaviour will be identical then. In the three compartments we will define the change in number of people with the following rules.

On the first day a specific number of people enter the first vaccinated compartment, while these are in the compartment and we are given the information that they are immunized only after the third day we know that these are prone to get infected until the third day, after which all the uninfected people remaining in this compartment will be transfered to the immunized compartment. But until the third day there will be some amount of people moving from the vaccinated to the infected compartment.

Similarly, we will implement these rules in the next two compartments as well and we realize that the dynamics of the compartments is just like the one compartment multiple dosage problem for each compartments and we obtain a curve referencing our intuition as well. For better comprehension of values the time axis for the readings have been a little shifted in the figure.

We also infer that the number of vaccinated and immunized will be almost similar but immunized will be always less than vaccinated because of this movement of vaccinated people into the infected compartment.

Coming to the duration and intensity of the epidemic

Duration of epidemic is until the time when no infected remains in the population(infected<1). Increasing the rate of vaccination will decrease the duration of epidemic because vaccination of more people will lead to a decrease in the number of susceptibles very quickly.

If the peak of the curve for number of infected people is recognized as the intensity of the epidemic then we see that increase in the vaccination rate will cause a decrease in the intensity of the epidemic. This phenomenon happens because the pool of susceptibles is decreased by immunization due to vaccination. Less number of people to get infected will cause the peak of the infected to be less than the previous values.
